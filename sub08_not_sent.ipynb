```python
train = pd.read_csv('train.csv')
train.head()
```

```python
%%time

feature_drop = [
    'feature_191','feature_192','feature_193','feature_194',
    'feature_195','feature_196','feature_197','feature_198',
    'feature_199','feature_200','feature_201','feature_202',
    'feature_203','feature_204']

train_ = train.drop(feature_drop, axis=1)

train_object = train_[train_.dtypes[train_.dtypes == 'object'].keys()]
feature_object = train_object.columns.tolist()

metric = []
for column in feature_object:
    contigency = pd.crosstab(train.target, train[column])
    c, p, dof, expected = chi2_contingency(contigency)
    metric.append(p)

metrica = pd.concat([pd.DataFrame(feature_object, columns=['variable']),pd.DataFrame(metric, columns=['metric'])], axis=1)
metrica = metrica.sort_values('metric', ascending=True)
#metrica.to_csv('metric.csv', sep=',', index = False)
```

```python
metrica_ = metrica[(metrica['metric'] > 0) & (metrica['metric'] < 0.0001)]
feature_object = metrica_.variable.to_list()
```

```python
train_object = train_object[feature_object]
train_object_ = pd.get_dummies(train_object, columns=feature_object,drop_first=True)
train_object_.head()
```
