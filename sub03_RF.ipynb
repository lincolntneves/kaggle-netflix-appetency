```python
import pandas as pd
import numpy as np
import os
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_predict
from sklearn.linear_model import LogisticRegression
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
## Dimensões para visualização de datasets
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.set_option('display.float_format', '{:.8f}'.format)
```

```python
train = pd.read_csv('train.csv')
train.head()
```

```python
train.shape
```

```python
train.info(verbose=True, null_counts = True)
```

```python
train_no_object = train[train.dtypes[train.dtypes != 'object'].keys()]
train_no_object.head()
```

```python
train_no_object.info(verbose=True, null_counts = True)
```

```python
X_train = train_no_object.set_index('id').fillna(train_no_object.mean()).drop(['target','feature_193','feature_196','feature_197','feature_198'], axis = 1)
X_train.head()
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X_train)
```

```python
X
```

```python
y = train['target']
```

```python
y
```

```python
%%time
clf_RF = RandomForestClassifier(random_state=12345)
clf_RF.fit(X,y)
```

## Predict test dataset with the first ML model

```python
test = pd.read_csv('test.csv')
test.head()
```

```python
test_no_object = test[test.dtypes[test.dtypes != 'object'].keys()]
test_no_object.head()
```

```python
test_ = test_no_object.set_index('id').fillna(test_no_object.mean()).drop(['feature_193','feature_196','feature_197','feature_198'], axis = 1)
test_.head()
```

```python
X_test = scaler.fit_transform(test_)
```

```python
X_test
```

```python
preds_RF = clf_RF.predict_proba(X_test)
```

```python
preds_RF_ = pd.DataFrame(preds_RF.T[1].T, columns=['target'])
```

```python
preds_RF_.head()
```

```python
sub03_RF = pd.concat([test['id'],preds_RF_], axis=1)
```

```python
sub03_RF.to_csv('sub03_LR.csv', sep=',', index=False)
```
